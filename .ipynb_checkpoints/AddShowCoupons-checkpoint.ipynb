# Present coupons

```python
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, Range1d, PanTool, WheelZoomTool, BoxSelectTool, Label, LabelSet
)
from bokeh.models.glyphs import Text

import requests

category_colors = {
                    'art': 'black', 
                    'cinema': 'blue', 
                    'travel': 'yellow', 
                    'parties': 'pink', 
                    'sport': 'red', 
                    'garden': 'green',
                    'groceries': 'orange',
                    'toys': '#B57EDC',
                    'clothes': '#89CFF0',
                    'shoes':'#483C32',
                    'electronic':'#00FF3F',
                    'motorization': 'grey',
                    'cosmetics': 'purple',
                    'home': 'brown',
                    'office': 'olive',
                    'construction': '#808080',
                    'shool': 'navy',
                    'babies': 'rose',
                    'concerts': '#BFFF00',
                    'theatre': 'peach',
                    'books': '#7F00FF',
                    'health': '#0095B6',
                    'pets': '#B87333',
                    'restaurants': '#A7FC00'
                    }

responseCo = requests.get(url='http://localhost:5000/coupon/getallcouponsdata')
allcouponsdata = responseCo.json()


responseFe = requests.get(url='http://localhost:5000/fence/alldatafences')
allfencesdata = responseFe.json()


f_names = [l['name'] for l in allfencesdata]
f_colors = [category_colors[l['category']] for l in allfencesdata]
f_radiuses = [float(l['radius']) for l in allfencesdata]
f_latitudes = [float(l['latitude']) for l in allfencesdata]
f_longitudes = [float(l['longitude']) for l in allfencesdata]



map_options = GMapOptions(lat=50.0614300, lng=19.9365800, map_type="roadmap", zoom=11)
plot = GMapPlot(x_range=Range1d(), y_range=Range1d(), map_options=map_options, plot_height=800, plot_width=980)
plot.title.text = "Krakw"
plot.api_key = "AIzaSyC2s9N7qsU48BL83Zaj-OCE6pqpvtTsxtM"




colors = [category_colors[l['category']] for l in allcouponsdata]
latitudes = [float(l['latitude']) for l in allcouponsdata]
longitudes = [float(l['longitude']) for l in allcouponsdata]
        


source_for_map = ColumnDataSource(
    data=dict(
        
        lat = latitudes,
        lon = longitudes,
        col = colors
    )
)

source_for_map_fences = ColumnDataSource(
    data=dict(
        names = f_names,
        rad = f_radiuses,
        lat = f_latitudes,
        lon = f_longitudes,
        col = f_colors
    )
)


circles = Circle(x="lon", y="lat", size=10, fill_color='col', line_color="black")
f_circles = Circle(x="lon", y="lat", radius="rad", fill_alpha=0, line_color="col")



plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())


plot.add_glyph(source_for_map, circles)
plot.add_glyph(source_for_map_fences, f_circles)

show(plot)
```
