```python
import requests
response = requests.get(url='http://localhost:5000/fence/alldatafences')
allfencesdata = response.json()
print(allfencesdata)
```

```python
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, Range1d, PanTool, WheelZoomTool, BoxSelectTool, Label, LabelSet
)
from bokeh.models.glyphs import Text




map_options = GMapOptions(lat=50.0614300, lng=19.9365800, map_type="roadmap", zoom=11)
plot = GMapPlot(x_range=Range1d(), y_range=Range1d(), map_options=map_options, plot_height=800, plot_width=980)
plot.title.text = "Kraków"
plot.api_key = "AIzaSyC2s9N7qsU48BL83Zaj-OCE6pqpvtTsxtM"



category_colors = {'undefined': 'black', 
                    'cinema': 'blue', 
                    'travel': 'yellow', 
                    'parties': 'pink', 
                    'sport': 'red', 
                    'garden': 'green',
                    'groceries': 'orange'}



fnames = [l['name'] for l in allfencesdata]
colors = [category_colors[l['category']] for l in allfencesdata]
radiuses = [float(l['radius']) for l in allfencesdata]
latitudes = [float(l['latitude']) for l in allfencesdata]
longitudes = [float(l['longitude']) for l in allfencesdata]
        


source_for_map = ColumnDataSource(
    data=dict(
        names = fnames,
        rad = radiuses,
        lat = latitudes,
        lon = longitudes,
        col = colors
    )
)


circles = Circle(x="lon", y="lat", radius="rad", fill_color='col', fill_alpha=0.5, line_color="blue")
labels = LabelSet(x="lon", y="lat", text='names', level='glyph',
              x_offset=0, y_offset=0, source=source_for_map, render_mode='canvas')

plot.add_glyph(source_for_map, circles)
plot.add_layout(labels)
plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())
show(plot)
```
